```python
import torch
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np
```

# Get imbalanced train set

```python
scal_train = torch.load(r"D:\Capstone\dataset\processed\CNN\CWRU\Scalogram\train.pt")
spec_train = torch.load(r"D:\Capstone\dataset\processed\CNN\CWRU\Spectrogram\train.pt")
```

# Get generated set

```python
def get_generated(path, cls):
    gen = torch.load(path, map_location=torch.device("cpu"))
    lab = torch.tensor([cls] * len(gen))
    return {"data": gen, "label": lab}
```

```python
scal_b = get_generated(r"D:\Capstone\dataset\generated\CWRU\ball_data.pt", 3)
scal_ir = get_generated(r"D:\Capstone\dataset\generated\CWRU\innerrace_data.pt", 2)
scal_n = get_generated(r"D:\Capstone\dataset\generated\CWRU\normal_data.pt", 0)
```

# Concatenate generated with train set

```python
train_new = {
    "data": torch.cat([scal_train["data"], scal_b["data"], scal_ir["data"], scal_n["data"]], dim=0),
    "label": torch.cat([scal_train["label"], scal_b["label"], scal_ir["label"], scal_n["label"]], dim=0)
}
```

```python
save_path = r"D:\Capstone\dataset\processed\CNN\CWRU\Scalogram\train_synthetic.pt"
torch.save(train_new, save_path)
```
