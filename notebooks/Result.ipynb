```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams.update({
    "font.family": "serif",
    "font.serif": ["Times New Roman"],
    "font.size": 10,

    "axes.labelsize": 10,
    "axes.titlesize": 10,
    "legend.fontsize": 8,

    "xtick.labelsize": 8,
    "ytick.labelsize": 8,

    "lines.linewidth": 1.2,

    "figure.dpi": 300,
    "savefig.dpi": 300,

    "axes.grid": True,
    "grid.linestyle": "--",
    "grid.linewidth": 0.5,
})
```

```python
files = [
    r"D:\Capstone\experiments\CNN\history_densenet121.csv",
    r"D:\Capstone\experiments\CNN\history_efficientnet_b0.csv",
    r"D:\Capstone\experiments\CNN\history_mobilenet_v3_large.csv",
    r"D:\Capstone\experiments\CNN\history_regnet_x_400mf.csv",
    r"D:\Capstone\experiments\CNN\history_shufflenet_v2_x1_5.csv"
]

dfs = [pd.read_csv(f) for f in files]
model_names = ["DenseNet-121", "EfficienNet-B0", "MobileNetV3-Large", "RegNetX-400mf", "ShuffleNetV2-X1.5"]
```

```python
def plot_metric_ieee(dfs, metric, model_names):
    plt.figure(figsize=(3.5, 2.5))  # IEEE column width

    for df, name in zip(dfs, model_names):
        plt.plot(df["epoch"], df[metric], label=name)

    plt.xlabel("Epoch")
    plt.ylabel(metric.replace("_", " ").title())
    plt.title(metric.replace("_", " ").title())

    plt.legend(frameon=False)
    plt.tight_layout()

    plt.savefig(f"{metric}_ieee.png", bbox_inches="tight")
    plt.show()
```

```python
metrics = ["train_loss", "train_acc", "val_loss", "val_acc"]

for m in metrics:
    plot_metric_ieee(dfs, m, model_names)
```
