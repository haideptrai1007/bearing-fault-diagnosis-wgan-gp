```python
import torch
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np
```

```python
train_old = torch.load(r"D:\Capstone\dataset\processed\TrainCNN\Scalogram\train.pt")
test_old = torch.load(r"D:\Capstone\dataset\processed\TrainCNN\Scalogram\test.pt")
val_old = torch.load(r"D:\Capstone\dataset\processed\TrainCNN\Scalogram\val.pt")
```

```python
def get_synthetic(path, cls):
    gen = torch.load(path, map_location=torch.device("cpu"))
    lab = torch.tensor([cls] * len(gen))
    return {"data": gen, "label": lab}
```

```python
b_dict = get_synthetic(r"D:\Capstone\dataset\synthetic\ball_data.pt", 3)
ir_dict = get_synthetic(r"D:\Capstone\dataset\synthetic\innerrace_data.pt", 2)
n_dict = get_synthetic(r"D:\Capstone\dataset\synthetic\normal_data.pt", 0)

print("B:", len(b_dict['label']))
print("IR:", len(ir_dict['label']))
print("N:", len(n_dict['label']))
```

```python
train_new = {
    "data": torch.cat([train_old["data"], b_dict["data"], ir_dict["data"], n_dict["data"]], dim=0),
    "label": torch.cat([train_old["label"], b_dict["label"], ir_dict["label"], n_dict["label"]], dim=0)
}
```

```python
save_path = r"D:\Capstone\dataset\processed\TrainCNN\Scalogram\train_synthetic.pt"
torch.save(train_new, save_path)
```
